# Part 2: Flow Resistance Calculation
Here, we provide the equation describing flow resistance in a rectangular microchannel. We have provided you the function for flow resistance (you do not need to modify this, just run the cells).

![alt text](images_lab2/resistance.png)


Figure from: Ainla, A. (2013). The Multifunctional Pipette. A Microfluidic Technology for the Biosciences.

```python
#you do not need to modify this cell, just run as is

#import packages
import numpy as np

###CONSTANTS - in SI units
#properties of liquid
#approximation for PBST w/ 1% dye 
viscosity = 10e-4  #Pa*s, approximation for 0.05% Tween 20 from Szymczyk et al., Colloids Interfaces, 2018
surfTension = 0.035  #N/meter, approximation for 0.05% Tween 20 from Rehman et al., J. Mater. Environ. Sci., 2017

#contact angle measurement, in radians
#hydrophobic top PDMS surface and hydrophilic side and bottom surfaces respectively
#changed side wall value due to using PBST w/ dye 
#approximated side wall angle using Fig S12, Shafique H, et al, Lab Chip, 2024.

thetaTopWall = (np.pi * 91.5/180) #contact angle of cover tape, in radians 
thetaSideWalls = (np.pi * 56/180) #approx. contact angle of PBST (no plasma treatment), in radians 

#width and height dimensions (placeholders)
w = 400e-6 #meters    
h = 400e-6 #meters 
l = 200e-3 #meters 
```

```python
#you do not need to modify this cell, just run as is
#FLOWRES
def flowResistance(w,h,l,viscosity):
    h_actual = min(w,h)
    w_actual= max(w,h)
    flowRes = ((12*viscosity*l)/(1-0.63*(h_actual/w_actual)))*(1/(w_actual*(h_actual**3)))
    return flowRes
```

## Call the flowResistance function using the default constants defined for you already. </font>

```python
#TO-DO: Call the flowResistance function using the default constants and variables. Print results.
flow_resistance = flowResistance(w, h, l, viscosity)
print("The flow resistance of the channel is:", flow_resistance, "PaÂ·s/mÂ³")
```

# Part 3: Capillary Pressure Function</font>

In the figure below is the Young-Laplace Equation describing the capillary pressure of a rectangular microchannel. P is the capillary pressure, Î³ is the surface tension of liquid in the microchannel. h and w are the channel height and width respectively. $Î¸_t$, $Î¸_b$, $Î¸_r$, $Î¸_l$, are the top, bottom, right, and left channel wall contact angles, respectively. 

Given the Young-Laplace Equation describing the capillary pressure of a rectangular microchannel, write the capPressure function.  The function should take in w, h, thetaTopWall, thetaSideWalls, and surfTension and return capPressure.  **Assume thetaTopWall as equivalent to $Î¸_t$, and thetaSideWalls as equivalent to $Î¸_b$, $Î¸_r$, and $Î¸_l$.**

    <img src="images_lab2/c8lc00458g-t1_hi-res.gif" width="600" />

    <img src="images_lab2/Screenshot_res.png" width="400" />

Adapted from Olanrewaju et al, 2018, Lab Chip,  https://doi.org/10/gk8xks and Olanrewaju et al, 2016, Lab Chip. https://doi.org/10.1039/C6LC00764C

```python
#CAPPRESSURE
def capPressure(w, h, thetaTopWall, thetaSideWalls,surfTension):
    #TO-DO: Capillary pressure calculation for wetting liquid
    #hint use "np.cos(angle)"to call a cosine function
    #fill in function here
    capPres = -surfTension * ( (np.cos(thetaTopWall) + np.cos(thetaSideWalls)) / h + (np.cos(thetaSideWalls) + np.cos(thetaSideWalls)) / w )
    return capPres
```

Call the capPressure function using the default constants defined for you already. </font>

```python
#TO-DO: Call the capPressure function using the default constants and variables. Print results.
cap_Pressure = capPressure(w, h, thetaTopWall, thetaSideWalls, surfTension)
print("The capillary pressure of the channel is:", cap_Pressure, "Pa")
```

# Part 4: Flow Rate Function</font>

Given that we have already defined functions for capillary pressure (P) and flow resistance (R) in a microchannel, we can now use these two values to define a flow rate (Q) function (ie. rearranging the formula below). </font>

<div style="text-align: center;">
    <img src="images_lab2/c8lc00458g-t7_hi-res.gif" width="200" />
</div>



```python
def flowRate(calc_cap_pres,totalFlowRes):
    #TO-DO: flowRate function
    flowrate = calc_cap_pres / totalFlowRes
    return flowrate
```

Call the flowrate function using the results from running the capPressure function and flowResistance function with the default variables.

```python
#TO-DO: Call the flowrate function using the default constants and variables. Print results.
flow_Rate = flowRate(cap_Pressure, flow_resistance)
print("The flow rate of the channel is:", flow_Rate, "mÂ³/s")
```

# Part 5: Volume and Drain Time Functions</font>

These are pretty self-explanatory. Using the resulting flow rate calculation and volume calculation, fill out the function for drain time. Note: if you have a negative flow rate value, this indicates that the flow is moving in the opposite direction of the reference direction, and you will need to account for this in your drainTime function (after all, time cannot be negative).

```python
#VOLUME
def Volume(w,h,l):
    V = w*h*l
    return V
```

```python
def drainTime(flowrate,vol):
    #TO-DO: Use the volume and flow rate to write a function that returns drain time (ie. time for the volume of the channel to fill).
    time = vol / np.abs(flowrate)
    return time
```

Call the Volume and drainTime functions with the default variables.  </font>

```python
#TO-DO: Call the volume and drainTime functions using the default variables. Print results.
drain_Time = drainTime(flow_Rate, Volume(w,h,l))
print("The drain time of the channel is:", drain_Time, "s")
```
